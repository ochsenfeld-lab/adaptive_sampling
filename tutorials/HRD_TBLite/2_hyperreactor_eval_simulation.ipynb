## Initialize NanoSim object

```python
# type 1 (w/o 'mols.dat' file)
nanoSim = NanoSim(traj_file="data/traj_compressed.traj", bond_order_file="data/compressed_bo.npz", pop_file="data/compressed_pop.npz")

# type 2 (w/ 'mols.dat' file) - if the mols file has already been generated with the above function
#nanoSim = NanoSim(traj_file="data/traj_compressed.traj", bond_order_file="data/compressed_bo.npz", pop_file="data/compressed_pop.npz", mols_file="data/mols.dat")

# compute fragments
# comment next row if you use second type of initialization
nanoSim.generate_frag_lists()
```

## Generate SMILES and Fill in Data Frame

```python
nanoSim.generate_mols_smiles()
nanoSim.generate_df()
```

## Plot Molecule Grid and Bar Plot

```python
plot_tools.generate_mol_grid(nanoSim.df)
plot_tools.generate_bar_plot(nanoSim.df)
```

## Create Reactions List

```python
# calc start_ts_index as (equil_time%period_time) / (dt * read_rate) + 500 fs (20th time step)
# here 40 + 20 because we have an offset of 1 ps (40 time steps) due to equilibration and we want to start 500 fs (20 time steps) after equilibration
reactions_list = construct_reactions_list(nanoSim.df, start_ts_index=60, period_ts_steps=80) # the period is 2 ps = 80 time steps
```

## Generate Network

```python
# generate network
if reactions_list==[]:
    print('No events found!')
else:
    nanoNetwork = NanoNetwork()
    nanoNetwork.create_network(reactions_list)

    # plot network graphs
    plot_tools.generate_network_grid(nanoNetwork)
    plot_tools.plot_static_network_kk(nanoNetwork)
    plot_tools.plot_static_network_shell(nanoNetwork)
```

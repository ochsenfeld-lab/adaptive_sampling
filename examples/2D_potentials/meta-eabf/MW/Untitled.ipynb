```python
import numpy as np
from adaptive_sampling.processing_tools import mbar
import matplotlib.pyplot as plt
```

```python
def combine_simulations(list_of_trajs: list, list_of_metafs: list):
    meta_f = np.concatenate(list_of_metafs, axis=1)
    traj_list = []
    for traj in list_of_trajs:
        traj_list += traj
    exp_U, frames_per_traj = mbar.build_boltzmann(
        [traj for traj in list_of_trajs], 
        meta_f, 
        equil_temp=300.0,
    )
    return exp_U, frames_per_traj
```

```python
# generate windows for all simulations
#   -> each simulation might use its own grid and ext_sigma
#   -> the bin width in the grid should always be chosen <= ext_sigma!
traj_list1, indices1, meta_f1 = mbar.get_windows(
    grid,              
    cv1,               
    la1, 
    ext_sigma,         
    equil_temp=300.0   
)
traj_list2, indices2, meta_f2 = mbar.get_windows(
    grid,               
    cv2,               
    la2, 
    ext_sigma,         
    equil_temp=300.0
)

# combine all meta_fs and traj_lists
all_metafs = np.concatenate([meta_f1, meta_f2, ])  
all_trajs = [
    *traj_list1, 
    *traj_list2, 
]           

# build boltzmann factors and run MBAR over combined data
exp_U, frames_per_traj = mbar.build_boltzmann(
    all_trajs, 
    all_metafs, 
    equil_temp=300.0,
)
weights = mbar.run_mbar(
    exp_U,
    frames_per_traj, 
    max_iter=10000,
    conv=1.0e-4,
    conv_errvec=1.0,
    outfreq=100,
    device='cpu',
)

# weights have to be assigned to correct data frame! 
cv_data_combined = np.concatenate([cv1[indices1], cv2[indices2], ])

# get pmf from combined data
#  -> the grid can again be different from before
#  -> reweighting of PMFs to a different CV can be achieved by using the corresponding trajectory
pmf,_ = mbar.pmf_from_weights(
    grid, 
    cv_data_combined, 
    weights,
    equil_temp=300.0,
)
```

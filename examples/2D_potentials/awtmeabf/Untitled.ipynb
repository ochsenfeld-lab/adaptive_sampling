```python
import numpy as np
from adaptive_sampling.processing_tools import mbar
import matplotlib.pyplot as plt
```

```python
traj1 = np.loadtxt("lower_bound/CV_traj.dat", skiprows=1)
traj2 = np.loadtxt("upper_bound/CV_traj.dat", skiprows=1)
```

```python
ext_sigma = 2.0
```

```python
minimum   = traj1[:,1].min()     
maximum   = traj1[:,1].max()
bin_width = 2.0

grid = np.arange(minimum, maximum, bin_width)
len(grid)
```

```python
cv1 = traj1[:,1]  # trajectory of collective variable
la1 = traj1[:,2]  # trajectory of extended system (for US: center of window)
u1  = traj1[:,3]

cv2 = traj2[:,1]  # trajectory of collective variable
la2 = traj2[:,2]  # trajectory of extended system
u2  = traj2[:,3]
```

```python
plt.plot(cv1)
plt.plot(cv2)
```

```python
traj1_list, indices1, meta_f1 = mbar.get_windows(grid, cv1, la1, ext_sigma, equil_temp=300.0)
dU1_list, _, _ = mbar.get_windows(grid, u1, la1, ext_sigma, equil_temp=300.0)
exp_U1, frames_per_traj1 = mbar.build_boltzmann(
    traj1_list, 
    meta_f1, 
    dU_list=dU1_list,
    equil_temp=300.0,
)
exp_U1_no, frames_per_traj1 = mbar.build_boltzmann(
    traj1_list, 
    meta_f1, 
    equil_temp=300.0,
)
```

```python
weights1 = mbar.run_mbar(
    exp_U1,
    frames_per_traj1, 
    max_iter=10000,
    conv=1.0e-7,
    conv_errvec=1.0,
    outfreq=100,
    device='cpu',
)
```

```python
weights1_no = mbar.run_mbar(
    exp_U1_no,
    frames_per_traj1, 
    max_iter=10000,
    conv=1.0e-7,
    conv_errvec=1.0,
    outfreq=100,
    device='cpu',
)
```

```python
traj2_list, indices2, meta_f2 = mbar.get_windows(grid, cv2, la2, ext_sigma, equil_temp=300.0)
dU2_list, _, _ = mbar.get_windows(grid, u2, la2, ext_sigma, equil_temp=300.0)
exp_U2, frames_per_traj2 = mbar.build_boltzmann(
    traj2_list, 
    meta_f2, 
    dU_list=dU2_list,
    equil_temp=300.0,
)
exp_U2_no, frames_per_traj2 = mbar.build_boltzmann(
    traj2_list, 
    meta_f2, 
    equil_temp=300.0,
)
```

```python
weights2 = mbar.run_mbar(
    exp_U2,
    frames_per_traj2, 
    max_iter=10000,
    conv=1.0e-7,
    conv_errvec=1.0,
    outfreq=100,
    device='cpu',
)
```

```python
weights2_no = mbar.run_mbar(
    exp_U2_no,
    frames_per_traj2, 
    max_iter=10000,
    conv=1.0e-7,
    conv_errvec=1.0,
    outfreq=100,
    device='cpu',
)
```

```python
pmf1,_ = mbar.pmf_from_weights(grid, cv1[indices1], weights1)
pmf2,_ = mbar.pmf_from_weights(grid, cv1[indices1], weights1_no)
pmf3,_ = mbar.pmf_from_weights(grid, cv2[indices2], weights2)
pmf4,_ = mbar.pmf_from_weights(grid, cv2[indices2], weights2_no)
pmf1 -= pmf1.min()
pmf2 -= pmf2.min()
pmf3 -= pmf3.min()
pmf4 -= pmf4.min()
```

```python
plt.plot(grid,pmf1,label="lower") 
plt.plot(grid,pmf2,label="lower")
```

```python
plt.plot(grid,pmf3,label="upper")
plt.plot(grid,pmf4,label="upper")
```
